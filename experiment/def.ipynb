```python
import requests
import random
import math

class GaodeGeo:
    def __init__(self):
        self.key = '7b1694e4e39645fe00c2ff3e2737b780'

    def requestApi(self, url):
        re = requests.get(url).json()
        return re

    # 根据经纬坐标获取地址等信息
    def getInverseGeoCode(self, location):
        url = f'https://restapi.amap.com/v3/geocode/regeo?parameters&key={self.key}&location={location}'
        json_data = self.requestApi(url)
        if json_data['status'] == '1':
            area= json_data['regeocode']['formatted_address']
            return area
        else:
            return '获取失败'

 
# base_log：经度基准点， 
# base_lat：维度基准点， 
# radius：距离基准点的半径
def generate_random_gps(base_log=None, base_lat=None, radius=None):
    radius_in_degrees = radius / 111300
    u = float(random.uniform(0.0, 1.0))
    v = float(random.uniform(0.0, 1.0))
    w = radius_in_degrees * math.sqrt(u)
    t = 2 * math.pi * v
    x = w * math.cos(t)
    y = w * math.sin(t)
    longitude = y + base_log
    latitude = x + base_lat
    loga = '%.6f' % longitude
    lata = '%.6f' % latitude
    return loga, lata



#老师想如何生成？
#我的方案：圆随机生
gd = GaodeGeo()
lst=[]
for i in range(20):
    longitude_, latitude_ = generate_random_gps(base_log=121.506337, base_lat=31.245105, radius=10000)
    longitude_# 经度
    latitude_ # 纬度
    string=[str( longitude_),str(latitude_)]
    place=','.join(string)
    area = gd.getInverseGeoCode(place)
    print(area)  
    lst.append(area)
    
```

```python
import requests
import json

def get_location_x_y(place):
    url = 'https://restapi.amap.com/v3/geocode/geo?parameters'
    parameters = {
        'key':'7b1694e4e39645fe00c2ff3e2737b780', 
        'address':'%s' % place
    }
    
    page_resource = requests.get(url,params=parameters)
    text = page_resource.text       
    data = json.loads(text)         
    location = data["geocodes"][0]['location']
    return location

for i in range(19):
    
    def route_planning():
       from_place = lst[i]
       from_location = get_location_x_y(from_place)

       to_place = lst[i+1]
       to_location = get_location_x_y(to_place)

       url = 'https://restapi.amap.com/v3/direction/driving'
       parameters = {
            'key': '7b1694e4e39645fe00c2ff3e2737b780',
            'origin': str(from_location),
            'destination': str(to_location)
            }
       response = requests.get(url, parameters)
       txt = response.text
       return txt 
   
biglst=[]
for i in range(19):
    if __name__ == '__main__':
        biglst.append(route_planning())
        print(route_planning())
```

```python
n=0
for i in biglst:
        txt_dic = json.loads(i)
        txt = txt_dic
        if txt['status'] == '1':
                dis = int(txt['route']['paths'][0]['distance'])#距离
                tim = int(txt['route']['paths'][0]['duration'])#时间
                ori=eval(txt['route']['origin'])#起点
                des=eval(txt['route']['destination'])#终点
                taxi_fee=eval(txt['route']['taxi_cost'])#预估出租车价格
        
                print('出发地 = {}  目的地 ={}'.format(lst[n],lst[n+1]))
                print('origin = {} , destination ={}'.format(ori,des))    
                print('distance = {}m'.format(dis))
                print('time = {}s'.format(tim))
                print('taxi_cost = {}元'.format(taxi_fee))
                print('***********************************************************************************************************************')
                n+=1

```
